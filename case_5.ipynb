```python
Q4="""Binary to Decimal & Octal to Hexadecimal Conversion: Develop a python program to convert binary to decimal, octal to hexadecimal"""
print(Q4)

def binary_to_decimal(binary_str):
    try:
        return int(binary_str, 2)
    except ValueError:
        print(f"Error: '{binary_str}' is not a valid binary number. Please enter only 0s and 1s.")
        return None

def octal_to_hexadecimal(octal_str):
    try:
        return hex(int(octal_str, 8))[2:].upper()
    except ValueError:
        print(f"Error: '{octal_str}' is not a valid octal number. Please enter digits from 0-7.")
        return None

# Input from user
binary = input("Enter a binary number: ")
octal = input("Enter an octal number: ")

# Function calls and output
decimal_result = binary_to_decimal(binary)
if decimal_result is not None:
    print("Decimal value of binary:", decimal_result)

hex_result = octal_to_hexadecimal(octal)
if hex_result is not None:
    print("Hexadecimal value of octal:", hex_result)

print("---------------------------------------------1")
```
